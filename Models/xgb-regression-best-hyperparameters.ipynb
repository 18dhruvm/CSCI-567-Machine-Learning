```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from math import sqrt
import xgboost as xgb
import time
from sklearn.metrics import classification_report, accuracy_score, mean_squared_error
from tabulate import tabulate
import seaborn as sns
import matplotlib.pyplot as plt


# Load the training data
train_data = pd.read_csv('/kaggle/input/finaldf/final_df.csv')

train_data.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(
    train_data.drop(['score', 'id'], axis=1),
    train_data['score'],
    test_size=0.2,
    random_state=42
)
```

```python
# Create an XGBoost Regressor
model = xgb.XGBRegressor(objective='reg:squarederror', seed=42, max_depth= 4, learning_rate= 0.11, n_estimators= 66)

# Train the model and Make predictions
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)

# Evaluate the model
rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
print(f'RMSE: {rmse:.4f}')

score_table = pd.DataFrame({'Expected Score': y_valid, 'Actual Score': y_pred})
print(tabulate(score_table, headers='keys', tablefmt='psql'))
```
